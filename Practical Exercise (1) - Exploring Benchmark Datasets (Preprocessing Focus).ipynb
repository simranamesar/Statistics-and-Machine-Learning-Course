### Dataset A (Supervised Learning – Classification)

**Dataset: Credit Card Fraud Detection <br>
https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud?search=Supervised+Learning+–+Classification** <br>

About Dataset:
Credit card companies are able to recognize fraudulent credit card transactions so that customers are not charged for items that they did not purchase.

```python
import pandas as pd
```

```python
cc_data = pd.read_csv("creditcard.csv")
cc_data.head()
```

```python
selected_features = ['Amount', 'Time', 'V2', 'V4', 'V11']
target = 'Class'

cc_subset = cc_data[selected_features + [target]]
cc_subset.head()
```

**Target (Label) Column 
Class <br>
0 - Legitimate transaction <br>
1 - Fraudulent transaction <br>
Binary classification (supervised learning) problem.**

**Why These Features Were Selected: <br>
Time -  transaction timing related to fraud <br>
Amount - fraudulent transactions often show abnormal amounts <br>
V2, V4, V11 - informative in fraud detection**

```python
cc_subset.info()
cc_subset.describe()
```

### Dataset B (Unsupervised Learning)

**Dataset: Market Segmentation in Insurance Unsupervised <br>
https://www.kaggle.com/datasets/jillanisofttech/market-segmentation-in-insurance-unsupervised/data** <br>

About Dataset: Dataset summarizes the usage behavior of about 9000 active credit cardholders during the last 6 months. The file is at a customer level with 18 behavioral variables.

```python
segment_data = pd.read_csv("Customer data.csv")
segment_data.head()
```

```python
selected_features = ['CUST_ID', 'BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES_TRX']

segment_subset = segment_data[selected_features]
segment_subset.head()
```

**Target (Label) Column: None**

**Why These Features Were Selected: <br>
CUST_ID - unique customer identifier<br>
BALANCE - customer account balance <br>
BALANCE_FREQUENCY - indicates how regularly the balance is updated<br>
PURCHASES_TRX - transaction activity level**

```python
segment_subset.info()
segment_subset.describe()
```
