**Minimum and Maximum (manual)**

```python
sorted_data = sorted(data)
print("Sorted data:", sorted_data)
```

**Median (manual)**

```python
n = len(sorted_data)

if n % 2 == 0:
    median = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
else:
    median = sorted_data[n//2]

print("Median:", median)
```

**Most frequent value using a dictionary**

```python
frequency = {}

for value in data:
    if value in frequency:
        frequency[value] += 1
    else:
        frequency[value] = 1

max_freq = max(frequency.values())

most_frequent = []
for key, value in frequency.items():
    if value == max_freq:
        most_frequent.append(key)

print("Most frequent value:", most_frequent)
```

### TASK 5 – SMALL FUNCTION

```python
def describe_data(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    
    # Mean
    total = 0
    for value in data:
        total += value
    mean_value = total / n
    
    # Median
    if n % 2 == 0:
        median = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
    else:
        median = sorted_data[n//2]
    
    return {
        "minimum": sorted_data[0],
        "maximum": sorted_data[-1],
        "mean": mean_value,
        "median": median
    }

result = describe_data(data)
print(result)
```
