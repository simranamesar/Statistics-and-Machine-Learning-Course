# Practical Part (5): Group Comparisons & Relationships

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
```

## Independent T-Test

**Synthetic Dataset**

```python
np.random.seed(42)
```

**Create two groups of exam scores:
Group A (mean ≈ 70)
Group B (mean ≈ 75)**

```python
groupA = np.random.normal(70, 8, 30)
groupB = np.random.normal(75, 8, 30)
```

***T test***

```python
tstat_value, p_value = ss.ttest_ind(groupA, groupB)
```

```python
print("Mean A:", groupA.mean())
print("Mean B:", groupB.mean())
print("t-statistic:", tstat_value)
print("p-value:", p_value)
```

## ANOVA

```python
trad = np.random.normal(70, 6, 30)
online = np.random.normal(75, 6, 30)
blended = np.random.normal(80, 6, 30)
```

```python
fstat_value, p_val = ss.f_oneway(trad, online, blended)
print("F statistic:", fstat_value)
print("p-value:", p_val)
```

## Correlation

```python
study_hours = np.random.uniform(1, 10, 40)
exam_scores = 50 + 5*study_hours + np.random.normal(0, 5, 40)
correlation, p_value = ss.pearsonr(study_hours, exam_scores)
print("Correlation:", correlation)
print("p-value:", p_value)
```
